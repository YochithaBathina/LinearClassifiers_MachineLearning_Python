# Pair Plot of Iris Data

```python
#importing libraries
import pandas as pa
import seaborn as sb
import sklearn.datasets
from sklearn.datasets import load_iris
```

```python
#loading iris data
data_iris=sb.load_dataset('iris')
data_iris.iloc[::5]
```

```python
#plotting iris data
sb.set_style("whitegrid")
plot= sb.pairplot(data_iris, hue= 'species',markers=["o", "s", "D"])
```

The above graphs are a 2D plot between different features of Iris data, from the above graph we can have the following analysis:
1. We have 6 unique graphs between sepal length, sepal width, petal length and petal width and the others are mirror images (Ignoring the diagonal graphs).
2. If we consider sepal_lenght vs sepal_width graph we can easily separate setosa flower species from non-setosa flowers but the versicolor and virginica flowers are difficult to separate. The major drawback in this plot is it has high bias and high variance.
3. By considering the plot petal_length vs petal_width, where we can easily separate the setosa flowers for versicolor and virginica flowers as there is low variance and low bias, if the petal_length<2  and 0<petal width<1 then we can define those flowers as setosa flowers, the same can be applicable for versicolor and virginica species of flowers.
